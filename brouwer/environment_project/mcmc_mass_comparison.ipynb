```python
#!/usr/bin/python

import numpy as np
import pyfits
import os
import environment_utils as envutils

from matplotlib import pyplot as plt
%matplotlib inline

from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import gridspec
from matplotlib import rc, rcParams

# Make use of TeX
rc('text',usetex=True)

# Change all fonts to 'Computer Modern'
rc('font',**{'family':'serif','serif':['Computer Modern']})
```

```python
envnames = ['Void', 'Sheet', 'Filament', 'Knot']
envcolors = ['red', 'green', 'blue', 'gold']

esdnames = ['esd_total', 'esd_cen', 'esd_sat', 'esd_host']
esdcolors = ['black', 'blue', 'red', 'purple']

massnames = ['Mcen', 'Msat']
masscolors = ['blue', 'red']

#plotlabels = ['Average ESD', 'Central', 'Satellite', 'Host', '2-halo term']
plotlabels = ['None', '2-halo', 'fc', '2-halo+fc']

filename1 = '/disks/shear10/brouwer_veersemeer/mcmc_output/environment_mcmc_output_auto_fc1' # No 2-halo, fixed fc
filename2 = '/disks/shear10/brouwer_veersemeer/mcmc_output/environment_mcmc_output_A2halo' # Free 2-halo, fixed fc
filename3 = '/disks/shear10/brouwer_veersemeer/mcmc_output/environment_mcmc_output_mstarweight_fcsep' # No 2-halo, free fc
filename4 = '/disks/shear10/brouwer_veersemeer/mcmc_output/environment_mcmc_output_fc_A2halo' # Free 2-halo, free fc

burn = 10000
#burn = 0
```

```python
# Import MCMC results
mcmc1, end, mcmcname, headername, mcmcmask = envutils.import_mcmc(filename1, burn)
mcmc2, end, mcmcname, headername, mcmcmask = envutils.import_mcmc(filename2, burn)
mcmc3, end, mcmcname, headername, mcmcmask = envutils.import_mcmc(filename3, burn)
mcmc4, end, mcmcname, headername, mcmcmask = envutils.import_mcmc(filename4, burn)
print 'Samples: Created = %i, Burned = %i, Used = %i'%(end, burn, len(mcmc1['Mcen1']))
print

# Import ESD profiles
esdfiles, covfile, inputparams, acfrac = envutils.read_header(headername)
print 'Input parameters:', np.sort(inputparams.keys())

# Importing the fitted ESD profiles
data_x, data_y, error_h, error_l = envutils.read_esdfiles(esdfiles)

plt.plot(acfrac)
```

```python
# Taking the median result of all samples after the burn-in phase
fsat, esd_fracs, esds_med, masses_med1, esds, masses, fracs = envutils.calc_esds_masses(inputparams, mcmc1, esdnames, massnames, envnames)
fsat, esd_fracs, esds_med, masses_med2, esds, masses, fracs = envutils.calc_esds_masses(inputparams, mcmc2, esdnames, massnames, envnames)
fsat, esd_fracs, esds_med, masses_med3, esds, masses, fracs = envutils.calc_esds_masses(inputparams, mcmc3, esdnames, massnames, envnames)
fsat, esd_fracs, esds_med, masses_med4, esds, masses, fracs = envutils.calc_esds_masses(inputparams, mcmc4, esdnames, massnames, envnames)


print
print esds
print
print masses
```

```python
# Create a plot of the masses

Nrows = 2
Ncolumns = 1

x = np.arange(len(envnames))

plt.figure(1, figsize=(10, 7))


# Plot Msat and Mcen



#mass_errors = np.array([abs(masses_med[:,:,1]-masses_med[:,:,x]) for x in [0, 2]]).T


for m in [1,0]:
    c=0
    Nsubplot = 100*Nrows+10*Ncolumns+m+1
    plt.subplot(Nsubplot)
    for masses_med in [masses_med1, masses_med2, masses_med3, masses_med4]:
        plt.plot(x, masses_med[m,:,1], ls='--', marker='o', label=plotlabels[c], color = envcolors[c])
    
        plt.fill_between(x, masses_med[m,:,0], masses_med[m,:,2], alpha=0.4, color = envcolors[c])
        c=c+1
        
    ylabel = r' %s (M$_{\odot}$)'%massnames[m]
    
    plt.yscale('log', fontsize=15)
    plt.ylabel(r'%s'%ylabel, fontsize=15)
    plt.tick_params(axis='both', which='major', labelsize=15)

    plt.xticks(np.arange(len(envnames)), envnames)

plt.legend(fontsize=15, loc='best')    
plotname = 'mcmc_mass_comparison.pdf'

plt.savefig(plotname, format='pdf')
print 'Written: ESD profile plot:', plotname
```
