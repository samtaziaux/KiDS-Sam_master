```python
#!/usr/bin/python

import numpy as np
import pyfits
import os
from matplotlib import pyplot as plt
%matplotlib inline

centering = 'BCG'
envs = ['Void', 'Sheet', 'Filament', 'Knot']

path_gamacats = '/data2/brouwer/MergedCatalogues'
gamacatname = '%s/ShearMergedCatalogueAll_sv0.8_shuffdeltaR.fits'%path_gamacats
```

```python
print 'Importing GAMA catalogue:', gamacatname
gamacat = pyfits.open(gamacatname)[1].data

lensIDlist = gamacat['ID']

# Importing angular seperation
angseplist = gamacat['AngSep%s'%centering]
angseplist[angseplist<=0] = 0.

# Importing and correction log(Mstar)
logmstarlist = gamacat['logmstar']
fluxscalelist = gamacat['fluxscale'] # Fluxscale, needed for stellar mass correction
corr_list = np.log10(fluxscalelist)# - 2*np.log10(h/0.7)
logmstarlist = logmstarlist + corr_list

obsmask = (fluxscalelist<500)&(logmstarlist>5)

logmstarlist = logmstarlist[obsmask]
mstarlist = 10**logmstarlist
angseplist = angseplist[obsmask]
envlist = gamacat['envS4'][obsmask]
zlist = gamacat['Z'][obsmask]
```

```python
for env in xrange(len(envs)):
    
    envname = envs[env]
    envmask = (envlist == env)
    
    zaverage = np.mean(zlist[envmask])
    mstaraverage = np.mean(mstarlist[envmask])

    print '%s: mean(Z) = %g, mean(M*) = %g'%(envname, zaverage, mstaraverage)    
```

```python
# Create histogram for the four environments
def create_histogram(obsname, obslist, nbins, envs, envlist, norm):
        
    histvals, histbins = np.histogram(obslist, nbins)
    hists = {}
    
    for env in xrange(len(envs)):
        envname = envs[env]
        envmask = (envlist == env)
        
        histvals, histbins = np.histogram(obslist[envmask], histbins, normed = norm)
        histcens = np.array([(histbins[i]+histbins[i+1])/2 for i in xrange(len(histvals))])
    
        hists[envname] = histvals
 
        # Show results
        plt.plot(histcens, histvals, '', ls='-', label=envs[env])
    
#    plt.xscale('log')
    plt.yscale('log')
    plt.legend()
    
    plt.xlabel(obsname)
    
    plt.show()
    plt.clf()
    
    return histbins, hists, histcens
```

```python
# Creating the angular seperation histogram
angsepbins, angsephists, angsephistcens = create_histogram('Angular separation (kpc)', angseplist, 100, envs, envlist, False)
```

```python
# Printing the angular separation histogram to a file

filename = 'Angsephist_envS4bin.txt'
with open(filename, 'w') as file:
    print >>file, '# AngSep%s center'%centering, '	', 'Ngals (Voids)', '	', 'Ngals (Sheets)', '	', 'Ngals (Filaments)', '	', 'Ngals (Knots)'

with open(filename, 'a') as file:
	for i in xrange(len(angsephistcens)):
		print >>file, angsephistcens[i], '	', angsephists['%s'%envs[0]][i], '	', angsephists['%s'%envs[1]][i], '	', angsephists['%s'%envs[2]][i], '	', angsephists['%s'%envs[3]][i]


print 'Written:', filename
```

```python
shufhist = {}

for env in xrange(len(envs)):
    logmstartot = float(np.sum(logmstarhists[envs[env]]))
    shufhist[envs[env]] = np.trunc(minhist*logmstartot)
```

```python
shuflist = np.zeros(len(logmstarlist)) # Length of the merged GAMA catalog

for env in xrange(len(envs)):
    envname = envs[env]
    envmask = (envlist == env)
    for bin in xrange(nbins):
        binmask = (logmstarbins[bin]<=logmstarlist) & (logmstarlist<logmstarbins[bin+1])
        totmask = envmask & binmask
        nlenses = shufhist[envs[env]][bin]
        print 'Env: %s, Bin: %i, Nlenses: %i'%(envname, bin, nlenses)
        if nlenses>0:
            randones = np.zeros(len(shuflist[totmask]))
            randnums = np.random.choice(len(shuflist[totmask]), nlenses, replace=False)
            randones[randnums] = np.ones(nlenses)
            shuflist[totmask] = randones
```
