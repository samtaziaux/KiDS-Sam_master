```python
#!/usr/bin/python

import numpy as np
import pyfits
import os
from matplotlib import pyplot as plt
%matplotlib inline


from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import gridspec
from matplotlib import rc, rcParams


centering = 'BCG'
envs = ['Void', 'Sheet', 'Filament', 'Knot']

path_gamacats = '/data2/brouwer/MergedCatalogues'
gamacatname = '%s/GAMACatalogue_1.0.fits'%path_gamacats
```

```python
# Importing GAMA catalogue
print 'Importing GAMA catalogue:', gamacatname
gamacat = pyfits.open(gamacatname)[1].data

lensIDlist = gamacat['ID']

# Importing angular seperation
angseplist = gamacat['AngSep%s'%centering]
angseplist[angseplist<=0] = 0.

# Importing and correction log(Mstar)
logmstarlist = gamacat['logmstar']
fluxscalelist = gamacat['fluxscale'] # Fluxscale, needed for stellar mass correction
corr_list = np.log10(fluxscalelist)# - 2*np.log10(h/0.7)
logmstarlist = logmstarlist + corr_list

# Applying a mask to the galaxies
obsmask = (fluxscalelist<500)&(logmstarlist>5)

logmstarlist = logmstarlist[obsmask]
mstarlist = 10**logmstarlist
angseplist = angseplist[obsmask]

envlist = gamacat['envS4'][obsmask]
zlist = gamacat['Z'][obsmask]
ranklist = gamacat['rank%s'%centering][obsmask]

print 'Imported: %i of %i galaxies'%(len(logmstarlist), len(lensIDlist))
```

```python
# Calculating average redshift, log(M*) and satellite fraction of the lens samples (needed for halo model)
for env in xrange(len(envs)):
    
    envname = envs[env]
    envmask = (envlist == env)
    
    zaverage = np.average(zlist[envmask])
    mstaraverage = np.average(mstarlist[envmask])

    satmask = (ranklist >= 2)
    lenses = float(len(ranklist[envmask]))
    sats = float(len(ranklist[satmask*envmask]))
    fsat = sats/lenses
    
    print '%s: mean(Z) = %g, mean(M*) = %g, fsat = %g'%(envname, zaverage, mstaraverage, fsat)    
```

```python
# Create any histogram for the four environments
def create_histogram(obsname, obslist, nbins, envs, envlist, style, norm, weightlist):

    obsmin = np.amin(obslist)
    obsmax = np.amax(obslist)

    if 'log' in style:    
        if obsmin <= 0.:
            obsmin = np.amin(obslist[obslist>0.])
        bins = np.logspace(np.log10(obsmin), np.log10(obsmax), nbins)
        bins = np.append(bins, obsmax)
        
        plt.xscale('log')
        plt.yscale('log')

    if 'lin' in style:
        bins = np.arange(obsmin, obsmax, abs(obsmax-obsmin)/nbins)
        bins = np.append(bins, obsmax)
    
    histvals, histbins = np.histogram(obslist, bins)
    hists = np.zeros([len(envs), nbins])
    
    for env in xrange(len(envs)):
        envmask = (envlist == env)
       
        if type(weightlist) == np.ndarray:
            weight = weightlist[envmask]
            print 'The weights are:', weight
        else:
            weight = np.ones(len(obslist[envmask]))
       
        histvals, histbins = np.histogram(obslist[envmask], histbins, normed = norm, weights = weight)
        histcens = np.array([(histbins[i]+histbins[i+1])/2 for i in xrange(len(histvals))])
        hists[env] = histvals
 
        # Show results
        plt.plot(histcens, histvals, '', ls='-', label=envs[env])
    
    plt.legend()
    
    plt.xlabel(obsname)
    
    plt.show()
    plt.clf()
    
    return histbins, hists, histcens
```

```python
# Printing data to a text file
def write_textfile(filename, datanames, data):

    # Create the file with header    
    with open(filename, 'w') as file:
        
        printline = '# '
        for n in xrange(len(datanames)):
            printline = '%s	%s'%(printline, datanames[n])
        print >>file, printline
    
    # Write the data to the file
    with open(filename, 'a') as file:
        
        for i in xrange(len(data[0])): # For each element ...
            printline = ''
            for n in xrange(len(datanames)): # ... in each row:
                printline = '%s	%s'%(printline, data[n, i]) # Put it in the row ...
            
            print >>file, printline # ... and print it
    
    print 'Written:', filename
```

```python
# Calculate the number of lenses in each logmstar bin
shufhists = np.zeros([len(envs), nbins])

for env in xrange(len(envs)):
    logmstartot = float(np.sum(logmstarhists[env])) # Number of lenses in each original logmstar bin
    shufhists[env] = np.trunc(minhist*logmstartot) # Number of lenses in the shuffled logmstar bin
```

```python
# Create the shuffled logmstar catalog
shuflist = np.zeros(len(logmstarlist)) # Length of the merged GAMA catalog

for env in xrange(len(envs)):
    envname = envs[env]
    envmask = (envlist == env)
    for b in xrange(nbins):
        binmask = (logmstarbins[b]<=logmstarlist) & (logmstarlist<logmstarbins[b+1])
        totmask = envmask & binmask
        nlenses = shufhists[env, b]
        #print 'Env: %s, Bin: %i, Nlenses: %i'%(envname, b, nlenses)
        if nlenses>0:
            randones = np.zeros(len(shuflist[totmask]))
            randnums = np.random.choice(len(shuflist[totmask]), nlenses, replace=False)
            randones[randnums] = np.ones(nlenses)
            shuflist[totmask] = randones
```
