```python
#!/usr/bin/python


""" Create "shuffled environments" that contain an equal number of random lenses inside each delta8 bin as the "real environments" """

import pyfits
import numpy as np
import sys
import os

from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import gridspec
from matplotlib import rc, rcParams

# Make use of TeX
rc('text',usetex=True)

# Change all fonts to 'Computer Modern'
rc('font',**{'family':'serif','serif':['Computer Modern']})

%matplotlib inline
```

```python
mergedcatdir = '/data2/brouwer/MergedCatalogues'
if os.path.isdir(mergedcatdir):
	pass
else:
	mergedcatdir = '/disks/shear10/brouwer_veersemeer/MergedCatalogues'

mergedcatname = '%s/GAMACatalogue_1.0.fits'%mergedcatdir
delta8catname = '%s/gama_environment/delta8.fits'%mergedcatdir

mergedcat = pyfits.open(mergedcatname)

galIDlist = mergedcat[1].data['ID'] # IDs of all galaxies
galZlist = mergedcat[1].data['Z'] # Central Z of the galaxy
delta8list = mergedcat[1].data['delta8'] # Local density of the galaxy
envlist = mergedcat[1].data['envS4'] # Environment of the galaxy (0 void, 1 sheet, 2 filament, 3 knot)
```

```python
deltamin = -1
deltamax = 15
print deltamin, deltamax

delta8bins = np.arange(deltamin, deltamax, 0.1)
histbins = np.append(delta8bins, deltamax)
print len(delta8bins), len(histbins)
```

```python
# Creating a dictionary containing the galIDs within each delta_r-bin
galID_binned = dict()

for b in xrange(len(delta8bins)-1):
    delta8mask = (delta8bins[b]<=delta8list)&(delta8list<=delta8bins[b+1])
    galID_binned[b]=galIDlist[delta8mask]
```

```python
# the histogram of the data
delta8_hist, histbins, patches = plt.hist(delta8list, histbins, facecolor='blue')
plt.ylabel(r'Number of galaxies', fontsize=15)
plt.xlabel(r'Local overdensity $\delta_8$', fontsize=15)
```

```python
plt.plot(delta8bins, delta8_hist)
plt.ylabel(r'Number of galaxies', fontsize=15)
plt.xlabel(r'Local overdensity $\delta_8$', fontsize=15)
```

```python
envmask = [(env<=envlist)&(envlist<(env+1)) for env in np.arange(4)]

delta8_hist_env = [[],[],[],[]]

for env in np.arange(4):
    x, histbins, patches = plt.hist(delta8list[envmask[env]], histbins)
    delta8_hist_env[env] = np.append(delta8_hist_env[env], x)

    delta8_hist = np.array(delta8_hist)
    delta8_hist_env = np.array(delta8_hist_env)
```

```python
for env in np.arange(4):
    plt.plot(delta8bins, delta8_hist_env[env])
plt.ylabel(r'Number of galaxies', fontsize=15)
plt.xlabel(r'Local overdensity $\delta_8$', fontsize=15)
```

```python
shuffled_nums = dict()

for b in xrange(len(delta8bins)):
    print b, delta8_hist[b], delta8_hist_env[0][b]
    shuffled_nums[b] = np.random.random_integers(0, delta8_hist[b], delta8_hist_env[0][b])-1
    print shuffled_nums[b]
    
```
