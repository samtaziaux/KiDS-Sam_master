```python
#!/usr/bin/python

import numpy as np
import pyfits
import os

from matplotlib import pyplot as plt
%matplotlib inline

envnames = ['Void', 'Sheet', 'Filament', 'Knot']
esdnames = ['esd_total', 'esd_central', 'esd_sat', 'esd_host']
esds = np.array([['%s%i'%(esd, i+1) for i in xrange(len(envnames))] for esd in esdnames])

print esds

filename = 'mcmc_output'
burn = 500
```

```python
# Importing mcmc results
def import_mcmc(filename):

    mcmcname = '%s.fits'%filename
    headername = '%s.hdr'%filename
    
    print 'Importing MCMC results:', mcmcname, 'and', headername
    mcmc = pyfits.open(mcmcname)[1].data
    
    mcmcmask = (mcmc['Msat1'] != 0)
    end = np.sum(mcmcmask)
    
    return mcmc, end, mcmcname, headername
```

```python
def read_header(headername):
    header = open(headername, 'r').read().split('\n')
    header = np.array([header[i].split(' ') for i in xrange(len(header))])
    
    for line in header:
        if line[0] == 'datafile':
            esdfiles = line[1].split(',')
        elif line[0] == 'covfile':
            covfile = line[1]
        
    return esdfiles, covfile
```

```python
def read_esdfiles(esdfiles):
    
    data = np.loadtxt(esdfiles[0]).T
    data_x = data[0]

    data_y = np.zeros(len(data_x))
    error_h = np.zeros(len(data_x))
    error_l = np.zeros(len(data_x))
    
    print 'Imported ESD profiles: %i'%len(esdfiles)
    
    for f in xrange(len(esdfiles)):
        # Load the text file containing the stacked profile
        data = np.loadtxt(esdfiles[f]).T
    
        bias = data[4]
        bias[bias==-999] = 1
    
        datax = data[0]
        datay = data[1]/bias
        datay[datay==-999] = np.nan
    
        errorh = (data[3])/bias # covariance error
        errorl = (data[3])/bias # covariance error
        errorh[errorh==-999] = np.nan
        errorl[errorl==-999] = np.nan
        
        
        data_y = np.vstack([data_y, datay])
        error_h = np.vstack([error_h, errorh])
        error_l = np.vstack([error_l, errorl])
        
    data_y = np.delete(data_y, 0, 0)
    error_h = np.delete(error_h, 0, 0)
    error_l = np.delete(error_l, 0, 0)
    
    return data_x, data_y, error_h, error_l
```

```python
# Import MCMC results
mcmc, end, mcmcname, headername = import_mcmc(filename)
print 'Samples: Created = %i, Burned = %i, Used = %i'%(end, burn, end-burn)

# Import ESD profiles
esdfiles, covfile = read_header(headername)
data_x, data_y, error_h, error_l = read_esdfiles(esdfiles)
```

```python
esds_av = np.array([[np.mean(mcmc[esds[e,i]][burn:end], 0) for e in xrange(len(esdnames))] for i in xrange(len(envnames))])

print esds_av
```

```python
from matplotlib import pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import gridspec
from matplotlib import rc, rcParams

# Make use of TeX
rc('text',usetex=True)

# Change all fonts to 'Computer Modern'
rc('font',**{'family':'serif','serif':['Computer Modern']})



# Define the labels for the plot
h=0.1
xlabel = r'radius R [kpc/h$_{%g}$]'%(h*100)
ylabel = r'ESD $\langle\Delta\Sigma\rangle$ [h$_{%g}$ M$_{\odot}$/pc$^2$]'%(h*100)


for i in xrange(len(envnames)):
    for e in xrange(len(esdnames)):
        plt.figure(1, figsize=(10, 8))
        plt.plot(data_x, esds_av[i,e], label=esdnames[e])
    
        (error_l[e])[(error_l[e])>=data_y[e]] = ((data_y[e][(error_l[e])>=data_y[e]])*0.9999999999)
    
        plt.errorbar(data_x, data_y[i], yerr=[error_l[i], error_h[i]], ls='', marker='o')
        plt.ylabel(r'%s'%ylabel,fontsize=15)
        plt.xlabel(r'%s'%xlabel,fontsize=15)
        
    plt.autoscale(enable=False, axis='both', tight=None)
    plt.axis([1e1,1e4,1e-1,1e4])
    plt.ylim(1e-1,1e4)
    
        
    plt.legend()
    
    plt.xscale('log')
    plt.yscale('log')

    plt.show()
    plt.clf
```
